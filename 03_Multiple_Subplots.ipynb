# Simple Line Plot

```python
x = np.arange(0.0, 2.0, 0.01)
x

#parameter 1 mengindikasian nilai awal, parameter 2 mengindikasikan nilai akhir, parameter 3 berfungsi menentukan interval
#nilai awal bersifat inklusif dan disertakan
#nilai akhir bersifat eksklusif dan hanya dijadikan sebagai pembatas
```

```python
s = np.sin(2 * np.pi * x) 
s
```

```python
fig, ax=plt.subplots() #OO Style
ax.plot(x, s)
ax.set(xlabel='nilai x', ylabel='nilai sin(x)', title='Visualisasi nilai sin')
ax.grid()
plt.show()

#terdapat 3 parameter, parameter xlabel beri nilai x, parameter ylabel beri nilai nilai sin(x), parameter ke 3 yaitu judul diberi nilai Visualisasi nilai sin
#ax.grid berguna untuk memunculkan grid dalam tabel
```

# Multiple Subplots

```python
x1 = np.linspace(0.0,5.0,100)
x2 = np.linspace(0.0,2.0,100)

y1 = np.cos(2*np.pi*x1)
y2 = np.cos(2*np.pi*x2)

#parameter 1 untuk nilai kecil, parameter 2 untuk nilai maksimal yang bersifat inklusif, parameter 3 jumlah data point yg akan di generate
#menampung hasil cos dari variabel yg di tampung
```

```python
x1
```

# Multiple Subplots Dengan OO Style

```python
fig, (ax1,ax2)=plt.subplots(2,1)
fig.suptitle('Multiple Subplots') #judul

ax1.plot(x1,y1,'ro-') #r berarti warna merah pada plotting
ax1.set_ylabel('Nilai cos(x1)')
ax2.plot(x2,y2,'g.-')
ax2.set_ylabel('Nilai cos(x2)')

ax2.set_xlabel('Nilai x')
plt.show()

#Karena plotting terdiri dari 2 buah axes, maka judulnya akan ditempatkan pada figure, bukan pada axes. 
```

# Multiple Subplots Dengan pyplot Style

```python
plt.subplot(2,1,1) 

#2 parameter pertama berfungsi untuk mengatur layout dan parameter ke-3 berfungsi untuk mengendalikan axes pertama
plt.plot(x1,y1,'ro-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai $cos(x_1)$')

plt.subplot(2,1,2) 
#2 parameter pertama berfungsi untuk mengatur layout dan parameter ke-3 berfungsi untuk mengendalikan axes kedua

plt.plot(x2,y2,'g.-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```
