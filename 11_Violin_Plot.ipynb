# Violin Plot

Import Modul dan Cek Versi

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

```python
plt.violinplot(data)

plt.title('Simple Violin plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
              showextrema=False, #digunakan untuk menampilkan garis tengah, nilai minimum dan maksimum
              showmeans=False, #memunculkan indikator untuk nilai mean
              showmedians=False, #memunculkan indikator untuk nilai median
              quantiles=None) 

plt.title('Violin plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

# Horisontal Violin Plot

```python
plt.violinplot(data, vert=False) #untuk menghasilkan horisontal violin plot

plt.title('Horisontal Violin plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

# Multiple Violin Plot

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data, 4]
```

```python
plt.violinplot(data)

plt.title('Multiple Violin plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
