# Pengaturan Format Pada Plot

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') #r berarti merah, o berarti dot.
plt.axis([0,6,0,20])
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
t 
```

```python
#CARA TIDAK EFISIEN
plt.plot(t,t,'r--') #di plot dengan warna merah dan hasil plottingnya berupa garis putus-putus
plt.plot(t,t**2,'bs') #merepresentasikan data kuadratik, untuk parameternya b merepresentasikan blue dan markernya berbentuk square
plt.plot(t,t**3,'g^') #merepresentasikan data pangkat tiga, untuk parameternya g merepresentasikan green dan markernya berbentuk segitiga
plt.show()
```

```python
#dapat juga dilakukan dengan satu pemanggilan fungsi plot()

plt.plot(t, t, 'r--',
         t, t**2, 'bs',
         t, t**3, 'g^') 
plt.show()
```

# Plotting dengan Keywords

```python
data={'a': np.arange(50), 'c': np.random.randint(0,50,50), 'd': np.random.randn(50)}
data['b'] = data['a'] + 10*np.random.randn(50)
data['d'] = np.abs(data['d'])*100

data
```

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
