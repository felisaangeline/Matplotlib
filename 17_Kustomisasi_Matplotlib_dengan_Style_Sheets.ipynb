Import Module dan Cek Versi

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

# Predefined Styles

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
x= np.linspace(1, 10, 15) #nilai terkecil satu, nilai terbesar 10, terdapat 15 bilangan

plt.plot(x, x, 'bs-', label='lin') #'lin' adalah linear
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

# Multiple Style

```python
plt.style.use(['seaborn', 'dark_background'])
```

```python
x= np.linspace(1, 10, 15) 

plt.plot(x, x, 'bs-', label='lin') 
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

# Temporary Styling

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0, 2* np.pi)), 'r-o')
plt.show()
```

```python
with plt.style.context( 'dark_background'): 
    plt.plot(np.sin(np.linspace(0, 2* np.pi)), 'r-o')
plt.show()

#style dark_background hanya temporary
```
