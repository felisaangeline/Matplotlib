# Contour Plot/Contour Chart

Import Modul dan Cek Versi

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

# Sample dataset

```python
n = np.linspace(-3.0, 3.0, 100) 

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

# Contour Plot

```python
plt.contour(x, y, z)
plt.colorbar()

plt.title('Contour plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

# Filled Contour Plot

```python
plt.contourf(x, y, z)
plt.colorbar()

plt.title('Filled Contour plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
